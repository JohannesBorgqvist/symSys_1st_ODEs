# Ok, so we decide to build our own function
This doesn't just cut it man! Let's build our own function! 

```python
# The functions substitutes the an unknown coefficient in the provided tangent
# with the provided output function. For some reason the standard subs function
# breaks down with very complicated integral expressions and it gives an error
# message saying "ValueError: substitution can not create dummy dependencies".
# So to circiumvent this we build our own function which conducts the substitution
# at hand
def substitute_tangents(old_tangent,original_function,target_function):
    # Allocate the new tangent which we will return
    new_tangent = old_tangent
    # Extract the coefficient
    coeff_temp = new_tangent.coeff(original_function)    
    # Check if the coefficient exist in the tangent
    if coeff_temp!=0:
        # Extract the coefficient
        coeff_temp = new_tangent.coeff(original_function)
        # Substract the original function and add the target!
        new_tangent += -(original_function*coeff_temp) + target_function*coeff_temp             
        # Expand and cancel things
        new_tangent =  cancel(expand(new_tangent))
    # Return the tangent at last
    return new_tangent
```
