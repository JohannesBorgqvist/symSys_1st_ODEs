# Error search ODE system and algebra system DBH model ansatz degree 1
*Date:* 2021-09-16,<br>
*Written by:* Johannes Borgqvist.<br>

This document contains an attempt of finding the error in the calculations of the symmetry generators for Hydon's model. The model at hand is the following two component ODE system:

\begin{align*}
\dfrac{\mathrm{d}w_{1}}{\mathrm{d}t}&=- w_{1} w_{2} - w_{1} w_{3} + w_{2} w_{3}=\omega_1(t,w_1,w_2,w_3)\\
\dfrac{\mathrm{d}w_{2}}{\mathrm{d}t}&=- w_{1} w_{2} + w_{1} w_{3} - w_{2} w_{3}=\omega_2(t,w_1,w_2,w_3)\\
\dfrac{\mathrm{d}w_{3}}{\mathrm{d}t}&=w_{1} w_{2} - w_{1} w_{3} - w_{2} w_{3}=\omega_3(t,w_1,w_2,w_3)\\
\end{align*}
To this model, the aim is to find the most general form of the *infinitesimal generator of the Lie group* denoted by $X$ which is defined as follows:

$$X=\xi(t,w_1,w_2,w_3)\partial_t+\eta_1(t,w_1,w_2,w_3)\partial_{w_1}+\eta_2(t,w_1,w_2,w_3)\partial_{w_2}+\eta_3(t,w_1,w_2,w_3)\partial_{w_3}.$$

To find this generator, a set of *linear ans채tze* is used for the three tangents as follows:

\begin{align*}                                                                                         
\xi&=w_{1} \operatorname{c_{0 1}}{\left(t \right)} + w_{2} \operatorname{c_{0 2}}{\left(t \right)} + w_
{3} \operatorname{c_{0 3}}{\left(t \right)} + \operatorname{c_{0 0}}{\left(t \right)}\\                
\eta_{1}&=w_{1} \operatorname{c_{1 1}}{\left(t \right)} + w_{2} \operatorname{c_{1 2}}{\left(t \right)}
 + w_{3} \operatorname{c_{1 3}}{\left(t \right)} + \operatorname{c_{1 0}}{\left(t \right)}\\           
\eta_{2}&=w_{1} \operatorname{c_{2 1}}{\left(t \right)} + w_{2} \operatorname{c_{2 2}}{\left(t \right)}
 + w_{3} \operatorname{c_{2 3}}{\left(t \right)} + \operatorname{c_{2 0}}{\left(t \right)}\\           
\eta_{3}&=w_{1} \operatorname{c_{3 1}}{\left(t \right)} + w_{2} \operatorname{c_{3 2}}{\left(t \right)}
 + w_{3} \operatorname{c_{3 3}}{\left(t \right)} + \operatorname{c_{3 0}}{\left(t \right)}\\           
\end{align*} 
The aim is to find the nine arbitrary functions $c_{ij}(t)$ for the two indices $i,j\in\left\{0,1,2\right\}$. The equations required in order to find these constants are given by the three *linearised symmetry conditions* given by

$$X^{(1)}(w_k'-\omega_k(t,y_1,y_2))=0,\;\;\textrm{for}\;\;k\in\left\{1,2,3\right\}.$$
Here, $X^{(1)}$ corresponds to the prolonged generator given by

$$X^{(1)}=X+\eta_1^{(1)}\partial_{w_1'}+\eta_2^{(1)}\partial_{w_2'}+\eta_3^{(1)}\partial_{w_3'}$$
where the prolonged tangents are given by the *prolongation formula*:
$$\eta_k^{(1)}=D_t\eta_k-w_k'D_t\xi,\;\;\textrm{for}\;\;k\in\left\{1,2,3\right\}$$
where the total derivative is defined as follows: $D_t=\partial_t+w_1'\partial_{w_1}+w_2'\partial_{w_2}+w_3'\partial_{w_3}$. 

What is nice about the DBH model is that it has at least three known generators, namely the *scaling generator* given by

$$X=w_1\partial_{w_1}+w_2\partial_{w_2}+w_3\partial_{w_3}$$
and also the translation generator
$$X=\partial_t+\partial_{w_1}+\partial_{w_2}+\partial_{w_3}.$$
Thus, we now when the algorithm performs correctly in this case as the above generator should be returned as an output. 

Moreover, plugging in these ans채tze into the linearised symmetry conditions will result in a linear system of equations which can be formulated on matrix form as follows:

$$A\dfrac{\mathrm{d}\mathbf{c}(t)}{\mathrm{d}t}=B\mathbf{c}(t)$$
where the vector $\mathbf{c}(t)\in\mathcal{C}(\mathbb{R}^{16})$ contains the nine arbitrary coefficients in the tangential ans채tze. Typically, the number of equations are much larger than the number of unknowns meaning that if $A,B\in\mathcal{C}\left(\mathbb{R}^{n\times m}\right)$ then $n\gg m$ (in this case $m=16$). After row reducing this system and simplifying it is (in the best of worlds) possible to write the system on the following form:

\begin{align}
\dfrac{\mathrm{d}\mathbf{c}(t)}{\mathrm{d}t}&=B\mathbf{c}(t),\\
B_{\mathrm{algebraic}}\mathbf{c}(t)&=\mathbf{0}.
\end{align}
The first ODE system is a quadratic ODE system which can be solved using the Jordan decomposition. That is if 
$$B=P^{-1}JP$$
then the solution to the ODE system is given by
$$\mathbf{c}(t)=P^{-1}e^{Jt}P\mathbf{c}_0$$
for some initial condition $\mathbf{c}_{0}$ composed of arbitrary integration constants. Then the solution of the system of ODEs is plugged in to the algebraic equations given by the second matrix equation above. This will result in certain algebraic equations that can simplify the results even further. 

Now, the problem is that certain generators are obtained that do not solve the linearised symmetry conditions. This implies that the implementation of the algorithm is wrong, as the methodology of ans채tze can never yield non-solutions. Therefore, the Hydon example will be used to see if the error is introduced in the solution of the ODE system or if it is when certain simplifications are made when the algebraic equations are solved. 

What will be done in the subsequent cells is that all matrices will be printed out and then we will try to track down the error.
