# New linear algebra approach starting with Hydon's model

*Written by:* Johannes Borgqvist <br>
*Date:* 2021-06-21<br>

So, one thing that has been bothering us, is that the algorithm we have devoloped, works approximately as if one were to do the calculations by hand. However, what I am beginning to realise is that it seems that for each case that I investigate, another little tweak to the original code must be added, and it would be much nicer and more efficient if we could use linear algebra here. 

Again, let's focus at Hydon's model:

\begin{align}
\dfrac{\mathrm{d}y_1}{\mathrm{d}t}&=\dfrac{ty_1 +y_2^2}{y_1 y_2-t^2},\\
\dfrac{\mathrm{d}y_2}{\mathrm{d}t}&=\dfrac{ty_2 +y_1^2}{y_1 y_2-t^2}.\\
\end{align}

Now, we now that the projective ans√§tze results in a linear system of the type

$$A\dfrac{\mathrm{d}\mathbf{c}}{\mathrm{d}t}=B\mathbf{c}$$
where $A,B\in\mathbf{R}^{m\times n}$ where $m\gg n$. 

Now the first step, is correct which is that the number of equations are reduced by constructing the matrix $M=[-A|B]$ and then finding a basis for $\mathrm{col}(M^T)$. This will result in a smaller matrix system of the same type. Then the question is if we could row-reduce this system in order to get a quadratic system of the following type 
$$\dfrac{\mathrm{d}\mathbf{c}}{\mathrm{d}t}=B\mathbf{c}$$
where all the algebraic equations are stored in the matrix $B_{\mathrm{algebraic}}$ and where the algebraic equations are given by:
$$B\mathbf{c}=\mathbf{0}.$$
